```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Titanic-Dataset.csv")
df.head()
```

```python
x2 = df.Pclass
y2 = df.Age
sns.lineplot(x=x2, y=y2, data=df, color='blue', linewidth=1, linestyle='--')
plt.title("Line Plot of Pclass vs Age")
plt.xlabel("Pclass")
plt.ylabel("Age")
plt.show()
```

```python
sns.scatterplot(x=x2, y=y2, data=df)
```

```python
z2 = df.Embarked
```

```python
sns.barplot(x=x2, y=y2, data=df, width=0.5, color='green')
```

```python
sns.boxplot(x= z2 , y = y2, data=df, width=0.5, color='orange')
```

```python
sns.histplot(df.Fare, kde=True, color='purple', bins=30, stat='density')
```

```python
sns.set_theme(style="whitegrid")
sns.barplot(x="Pclass", y="Survived", data=df, palette="Blues_d")
sns.catplot(x="Pclass", y="Survived", data=df, kind="bar", palette="Blues_d")
sns.catplot(x="Pclass", y="Survived", data=df, kind="point", palette="Blues_d")
sns.catplot(x="Pclass", y="Survived", data=df, kind="strip", palette="Blues_d")
```

```python
sns.kdeplot(df.Fare)
```

```python
# heatmap 
cor = df.corr(numeric_only=True)
sns.heatmap(cor, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
